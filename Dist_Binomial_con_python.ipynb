```
#Importamos librerias de la siguiente forma: 
import numpy as np # importando numpy
from scipy import stats # importando scipy.stats
import matplotlib.pyplot as plt #importando matplotlib
import pandas as pd #pandas
```

##Ejercicio 1
El jefe de recursos humanos de una empresa realiza un test de diez 칤tems a los aspirantes a un puesto, teniendo en cada 칤tems cuatro posibles respuestas, de las que solo una es correcta. Suponiendo que los aspirantes tienen la misma probabilidad de responder. 
Se pide hallar las probabilidades para el aspirante, que conteste:

* a)  Todos los 칤tems mal 
* b)  Al menos cuatro 칤tems correctos 
* c)  Entre cuatro y seis 칤tems correctos 
* d)  Todos los 칤tems correctos 
* e)  Menos de tres 칤tems correctos


```
# Definir parametros 洧녦~洧냣(n, p) 

n = 10
p = 1/4

#Crear distribucion binomial
binomial = stats.binom(n, p) # Distribuci칩n binomial con Scipy
x = np.arange(binomial.ppf(0.000001), #creamos conjunto de valores para x con el inverso de cdf (cumulative dist. function)
              binomial.ppf(0.999999))
print(x)
fmp = binomial.pmf(x) #pmf: Funci칩n de Masa de Probabilidad

#graficamos distribucion
plt.plot(x, fmp, '--')
plt.vlines(x, 0, fmp, colors='b', lw=10, alpha=0.5)
plt.title('Distribuci칩n Binomial')
plt.ylabel('probabilidad')
plt.xlabel('valores')
plt.show()
```

```
#ahora calcular las probabilidades para el aspirante

#a) Todos los 칤tems mal 
cero = stats.binom(n, p).pmf(0) #usar pmf: Funci칩n de Masa de Probabilidad
print ('prob 0 correcta: ', cero)
```

```
#a) Todos los 칤tems mal 
cero = stats.binom(n, p).pmf(0) #usar pmf: Funci칩n de Masa de Probabilidad
print ('prob 0 correcta: ', cero)

#b) Al menos cuatro 칤tems correctos 
prob_4_o_mas = 1 - stats.binom(n, p).cdf(3) #usar cdf: Funcion de dist. acumulada
print('prob 4 o mas: ', prob_4_o_mas)

#c) Entre cuatro y seis 칤tems correctos 
prob_4_a_6 = stats.binom(n, p).pmf(4) + stats.binom(n, p).pmf(5) + stats.binom(n, p).pmf(6)
print('prob entre 4 y 6: ', prob_4_a_6)

#otra forma con sustracci칩n de prob
prob_4_a_6b =  stats.binom(n, p).cdf(6)-stats.binom(n, p).cdf(3)
print('prob entre 4 y 6*: ', prob_4_a_6b)

```

```
#d) Todos los 칤tems correctos 
prob_10 = stats.binom(n, p).pmf(10)#usar pmf: Funci칩n de Masa de Probabilidad
print('prob todos: ', prob_10)

#e) Menos de tres 칤tems correctos
prob_2_o_menos = stats.binom(n, p).cdf(2) #usar cdf: Funcion de dist. acumulada
print('prob 2 o menos: ', prob_2_o_menos)
```

##Ejercicio 2 
En cada una de las diferentes competencias, Antonia tiene un 60% de posibilidades de ganar. Suponiendo que las competiciones sean independientes entre s칤, 

* a) 쯖u치l es la probabilidad de que Antonia gane 1 carrera de 4?

* b) 쯖u치l es la probabilidad de que Antonia gane al menos 1 carrera de 4?

* c) 쯖u치l es la probabilidad de que Antonia gane al menos 1 carrera de 10?

```
# Definir parametros 洧녦~洧냣(n, p) 

n = 4
p = 0.6

#Crear distribucion
binomial = stats.binom(n, p) # Distribuci칩n binomial con Scipy
x = np.arange(binomial.ppf(0.0001), #creamos conjunto de valores para x con el inverso de cdf (cumulative dist. function)
              binomial.ppf(0.9999))
print(x)
fmp = binomial.pmf(x) #pmf: Funci칩n de Masa de Probabilidad

#graficamos distribucion
plt.plot(x, fmp, '--')
plt.vlines(x, 0, fmp, colors='b', lw=10, alpha=0.5)
plt.title('Distribuci칩n Binomial')
plt.ylabel('probabilidad')
plt.xlabel('valores')
plt.show()
```

```
#ahora calcular las probabilidades para Antonia

#a) 쯖u치l es la probabilidad de que Antonia gane 1 carrera de 4?
una = stats.binom(n, p).pmf(1) #usar pmf: Funci칩n de Masa de Probabilidad
print ('prob gane 1 carrera de 4: ', una)
```

```
#b) 쯖u치l es la probabilidad de que Antonia gane al menos 1 carrera de 4?
una_o_mas = 1 - stats.binom(n, p).pmf(0)  #usar pmf: Funci칩n de Masa de Probabilidad
print ('prob gane al menos 1 de 4: ', una_o_mas)
```

```
#c) 쯖u치l es la probabilidad de que Antonia gane al menos 1 carrera de 10?

n=10 #aqui cambia el numero de ensayos

una_o_mas_de_10 = 1 - stats.binom(n, p).pmf(0)  #usar pmf: Funci칩n de Masa de Probabilidad
print ('prob gane al menos 1 de 10: ', una_o_mas_de_10)
```
